```python
import matplotlib.pyplot as plt
import pandas as pd
import sys

# Check if file exists
try:
    df = pd.read_csv('results.csv')
except FileNotFoundError:
    print("Error: 'results.csv' not found. Please run the C program first.")
    sys.exit(1)

plt.figure(figsize=(10, 7))

# Plotting each column with unique markers
plt.plot(df['Size'], df['RowMajor'], marker='o', linewidth=2, label='Row Major (Standard)')
plt.plot(df['Size'], df['ColMajor'], marker='x', linewidth=2, linestyle='--', color='red', label='Col Major (Cache Misses)')
plt.plot(df['Size'], df['1DFlat'],   marker='s', linewidth=2, label='1D Flat')
plt.plot(df['Size'], df['Pointer'],  marker='^', linewidth=2, label='Pointer Arithmetic')
plt.plot(df['Size'], df['Blocked'],  marker='d', linewidth=2, label='Blocked (Tiled)')

# Formatting
plt.title('Matrix Addition Performance: Access Pattern Comparison')
plt.xlabel('Matrix Dimension (N)')
plt.ylabel('Execution Time (seconds) - Log Scale')

# Log Scale is required to see the difference between the fast methods
# vs the very slow Column Major method.
plt.yscale('log')

plt.grid(True, which="both", ls="-", alpha=0.3)
plt.legend()
plt.tight_layout()

# Save and Show
plt.savefig('matrix_comparison_plot.png')
print("Plot saved as 'matrix_comparison_plot.png'")
plt.show()
```
